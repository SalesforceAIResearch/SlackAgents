## Step 1: write a manifest json file


Following the instruction about [manifest](https://api.slack.com/reference/manifests), we can write a manifest json file. We put a example manifest json file in the [manifest](./example_manifest.json). You could modify the name as your agent name. Then we load the manifest json file as a python dictionary.

```python

import json
import pprint
manifest_path = "example_manifest.json"
manifest = json.load(open(manifest_path))
pprint.pprint(manifest)
```

## Step 2: create a Slack app with the manifest

There are multiple [methods](https://api.slack.com/reference/manifests#manifest_apis) to modify the manifest of the app including
- `apps.manifest.create` is used to create apps from JSON manifests.
- `apps.manifest.update` will let you update configurations of your existing apps.
- `apps.manifest.delete` can delete any of your apps.
- `apps.manifest.export` exports the manifest of your existing apps.
- `apps.manifest.validate` will validate any JSON manifest against the correct schema.

We will use `apps.manifest.create` to create a Slack app with the manifest.

First, you need to get the app configuration token as in [managing configuration tokens](https://api.slack.com/reference/manifests#config-tokens). Then you run the following block

```python
!export SLACK_APP_CONFIG_TOKEN=<your_app_config_token>
```

Then we could create the app with the manifest file by running the following block:

```python
import os
from slack_sdk import WebClient
token = os.getenv("SLACK_APP_CONFIG_TOKEN")
client = WebClient(token=token)
response = client.apps_manifest_create(manifest=manifest)
app_id = response["app_id"] # get the app id
```

Now you can find the app in your Slack workspace on the [Slack App Dashboard](https://api.slack.com/apps)

## Step 3: update the manifest of the app


You can also update the manifest of the app. Let's update the display name of the app to be "My_New_App".


```python
new_name = "My_New_App"
manifest["display_information"]["name"] = new_name
manifest["features"]["bot_user"]["display_name"] = new_name
response = client.apps_manifest_update(app_id=app_id, manifest=manifest)
```

Now you can find the app in your Slack workspace on the [Slack App Dashboard](https://api.slack.com/apps) and see the display name has been updated.
