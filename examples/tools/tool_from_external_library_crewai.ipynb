## Creating a Custom CrewAI tool
Now, let's create a simple multiplication tool using CrewAI's `BaseTool` class:


````python
from crewai_tools import BaseTool


class MultiplicationTool(BaseTool):
    name: str = "Multiplication Tool"
    description: str = "Use this tool to multiply two numbers."

    def _run(self, a: int, b: int) -> int:
        return a * b
````

## Use the Custom CrewAI Tool in SlackAgents
Now, let's wrap the CrewAI tool in a SlackAgents tool and use it in a SlackAgents agent.


````python
from slackagents.tools.base import ToolCall
from slackagents.tools.wrappers.crewai import CrewAIToolWrapper
import json

crewai_multiply_tool = MultiplicationTool()

crewai_multiply_tool.run(a=2, b=3)

tool = CrewAIToolWrapper(tool=crewai_multiply_tool)
print(json.dumps(tool.info, indent=4))

tool_call = ToolCall(
    name="Multiplication Tool",
    arguments={"a": 2, "b": 3},
)

result = tool.execute(tool_call)
print(f"Tool result: {result}")
````

## Explanation
In this example, we've created a simple multiplication tool using CrewAI and integrated it with SlackAgents. Here's what's happening:

1. We define a `MultiplyTool` class in CrewAI, which takes two numbers and returns their product.
2. We create an instance of this tool: `crewai_multiply_tool`.
3. We use the `CrewAIToolWrapper` class from SlackAgents to wrap our CrewAI tool, making it compatible with SlackAgents.
4. We print out the tool's info, which shows how it will be presented to the AI agent.
5. Finally, we execute the tool with sample inputs (2 and 3) and print the result.

This demonstrates how easily you can integrate CrewAI tools with SlackAgents, allowing you to leverage the capabilities of both libraries in your projects.

## Using Public CrewAI Toolkit in SlackAgents
SlackAgents also supports using public [CrewAI toolkits](https://github.com/crewAIInc/crewAI-tools/tree/main/crewai_tools) in your agents. Let's see how to use the `ScrapeWebsiteTool` in a SlackAgents agent.
    

````python
# To enable the tool to search any website the agent comes across or learns about during its operation
from crewai_tools import ScrapeWebsiteTool
# To enable scrapping any website it finds during it's execution
crewai_webscraping_tool = ScrapeWebsiteTool()
# Initialize the tool with the website URL, so the agent can only scrap the content of the specified website
crewai_webscraping_tool = ScrapeWebsiteTool(website_url='https://www.salesforce.com')

# Wrap the CrewAI tool in a SlackAgents tool
tool = CrewAIToolWrapper(crewai_webscraping_tool)

tool_call = ToolCall(
    name="ScrapeWebsiteTool",
    arguments={},
)

result = tool.execute(tool_call)
print(f"Tool result: {result}")
````
